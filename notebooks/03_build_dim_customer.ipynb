# 03 Build dim_customer

This notebook creates the customer dimension table from the cleaned sales dataset.

Each row in `dim_customer` represents one unique customer.

```python
import pandas as pd
import os
```

```python
input_path = "../data/processed/clean_sales.csv"
output_path = "../data/processed/dim_customer.csv"
```

```python
df = pd.read_csv(input_path)
df.head()
```

```python
df[["customer_id", "customer_name", "segment"]].head()
```

```python
dim_customer = df[["customer_id", "customer_name", "segment"]].copy()
dim_customer = dim_customer.drop_duplicates()
```

```python
print("Rows:", len(dim_customer))
print("Unique customer_id:", dim_customer["customer_id"].nunique())
print("Unique customer_name:", dim_customer["customer_name"].nunique())
```

```python
dim_customer[dim_customer.duplicated(subset=["customer_id"], keep=False)] \
    .sort_values("customer_id")
```

```python
dim_customer = (
    dim_customer
    .sort_values("customer_id")
    .drop_duplicates(subset=["customer_id"], keep="first")
    .reset_index(drop=True)
)
```

```python
text_cols = ["customer_id", "customer_name", "segment"]

for col in text_cols:
    dim_customer[col] = dim_customer[col].astype(str).str.strip()
```

```python
print("Rows:", len(dim_customer))
print("Unique customer_id:", dim_customer["customer_id"].nunique())
print("Unique customer_name:", dim_customer["customer_name"].nunique())
print("Segments:", dim_customer["segment"].unique())
```

```python
os.makedirs("../data/processed", exist_ok=True)

dim_customer.to_csv(output_path, index=False)

print(f"dim_customer saved to: {output_path}")
```
