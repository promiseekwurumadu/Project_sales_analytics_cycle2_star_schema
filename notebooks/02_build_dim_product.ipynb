### 02 Build dim_product

#### This notebook creates the product dimension table from the cleaned sales dataset.

#### Each row in `dim_product` represents one unique product.

```python
import pandas as pd
import os
```

```python
input_path = "../data/processed/clean_sales.csv"
output_path = "../data/processed/dim_product.csv"
```

```python
df = pd.read_csv(input_path)
df.head()
```

```python
df[["product_id", "product_name", "category", "sub_category"]].head()
```

```python
dim_product = df[["product_id", "product_name", "category", "sub_category"]].copy()
```

```python
dim_product = dim_product.drop_duplicates()
```

```python
print("dim_product shape:", dim_product.shape)
dim_product.head()
```

```python
dim_product.isna().sum()
```

```python
text_cols = ["product_name", "category", "sub_category"]

for col in text_cols:
    dim_product[col] = dim_product[col].astype(str).str.strip()
```

```python
dim_product = dim_product.sort_values(by=["category", "sub_category", "product_name"]).reset_index(drop=True)
```

```python
dim_product = dim_product.sort_values(by=["category", "sub_category", "product_name"]).reset_index(drop=True)
```

```python
print("Rows:", len(dim_product))
print("Unique product_id:", dim_product["product_id"].nunique())
print("Unique product_name:", dim_product["product_name"].nunique())
```

```python
os.makedirs("../data/processed", exist_ok=True)
dim_product.to_csv(output_path, index=False)
print(f"dim_product saved to: {output_path}")
```

```python
dim_product = (
    df[["product_id", "product_name", "category", "sub_category"]]
    .drop_duplicates()
    .sort_values("product_id")
)

# Keep first occurrence per product_id
dim_product = dim_product.drop_duplicates(subset=["product_id"], keep="first")
```

```python
print("Rows:", len(dim_product))
print("Unique product_id:", dim_product["product_id"].nunique())
```
