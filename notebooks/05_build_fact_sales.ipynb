# 05 Build fact_sales

This notebook creates the fact table from the cleaned sales dataset.

Each row in `fact_sales` represents one sales transaction line.

```python
import pandas as pd
import os
```

```python
input_path = "../data/processed/clean_sales.csv"
output_path = "../data/processed/fact_sales.csv"
```

```python
df = pd.read_csv(input_path)
df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["date_id"] = df["order_date"].dt.date
```

```python
fact_sales = df[[
    "date_id",
    "customer_id",
    "product_id",
    "quantity",
    "sales",
    "discount",
    "profit"
]].copy()
```

```python
fact_sales.insert(0, "sales_id", range(1, len(fact_sales) + 1))
```

```python
fact_sales.head()
```

```python
fact_sales.isna().sum()
```

```python
print("Rows:", len(fact_sales))
print("Unique sales_id:", fact_sales["sales_id"].nunique())
print("Unique customers:", fact_sales["customer_id"].nunique())
print("Unique products:", fact_sales["product_id"].nunique())
print("Date range:", fact_sales["date_id"].min(), "to", fact_sales["date_id"].max())
```

```python
os.makedirs("../data/processed", exist_ok=True)

fact_sales.to_csv(output_path, index=False)

print(f"fact_sales saved to: {output_path}")
```
