# 04 Build dim_date

This notebook creates the date dimension table from the cleaned sales dataset.

Each row in `dim_date` represents one calendar date.

```python
import pandas as pd
import os
```

```python
input_path = "../data/processed/clean_sales.csv"
output_path = "../data/processed/dim_date.csv"
```

```python
df = pd.read_csv(input_path)
df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
```

```python
print(df["order_date"])
```

```python
start_date = df["order_date"].min()
end_date = df["order_date"].max()

print("Start date:", start_date)
print("End date:", end_date)
```

```python
date_range = pd.date_range(start=start_date, end=end_date, freq="D")
```

```python
dim_date = pd.DataFrame({
    "date_id": date_range.date,
    "date": date_range.date,
    "year": date_range.year,
    "quarter": date_range.quarter,
    "month": date_range.month,
    "month_name": date_range.month_name(),
    "day": date_range.day,
    "day_name": date_range.day_name(),
    "week_of_year": date_range.isocalendar().week.astype(int),
    "is_weekend": date_range.dayofweek >= 5
})
```

```python
dim_date.head()
```

```python
print("Rows:", len(dim_date))
print("Unique date_id:", dim_date["date_id"].nunique())
print("Start:", dim_date["date"].min())
print("End:", dim_date["date"].max())
```

```python
os.makedirs("../data/processed", exist_ok=True)

dim_date.to_csv(output_path, index=False)

print(f"dim_date saved to: {output_path}")
```
